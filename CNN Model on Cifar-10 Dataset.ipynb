### *My first project of Computer Vision* : <u>CNN Model on CIFAR-10 dataset.</u>

```python
import tensorflow as tf
from tensorflow.keras import layers,models,datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test) = datasets.cifar10.load_data()
```

```python
(x_train.shape,y_train.shape),(x_test.shape,y_test.shape)
```

```python
for i in range(5):
    plt.imshow(x_train[i])
    plt.show()
```

```python
for i in range(5):
    plt.imshow(x_test[i])
    plt.show()
```

#### CNN with single conv2d,pooling layer and filter size = (2 x 2)

```python
model = models.Sequential([
    layers.Conv2D(filters = 53,activation='relu',input_shape=(32,32,3),kernel_size=(2,2)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(56,activation= 'relu'),
    layers.Dense(10,activation = 'softmax')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
x_train = x_train / 255
y_train = y_train / 255
x_test = x_test / 255
y_test = y_test / 255
```

```python
y_train = y_train*255
y_test = y_test*255
```

```python
model.fit(x_train,y_train, epochs = 10)
```

```python
model.evaluate(x_test,y_test)
```

#### Model with single Conv2D layer and kernel size(filter size) == (3 x 3)

```python
model_with_kernel_3 = models.Sequential([
    layers.Conv2D(filters = 53,activation='relu',input_shape=(32,32,3),kernel_size=(3,3)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(56,activation= 'relu'),
    layers.Dense(10,activation = 'softmax')
])
```

```python
model_with_kernel_3.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model_with_kernel_3.fit(x_train,y_train, epochs = 10)
```

```python
model_with_kernel_3.evaluate(x_test,y_test)
```

#### Model with 3 Conv2d, pooling layers and filter size = (3 x 3)

```python
model_with_kernel_3_and_3_conv2d = models.Sequential([
    layers.Conv2D(filters = 128,activation='relu',input_shape=(32,32,3),kernel_size=(3,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters = 128,activation='relu',input_shape=(32,32,3),kernel_size=(3,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters = 128,activation='relu',input_shape=(32,32,3),kernel_size=(3,3)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(56,activation= 'relu'),
    layers.Dense(10,activation = 'softmax')
])
```

```python
model_with_kernel_3_and_3_conv2d.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model_with_kernel_3_and_3_conv2d.fit(x_train,y_train, epochs = 10)
```

```python
model_with_kernel_3_and_3_conv2d.evaluate(x_test,y_test)
```

```python
model_with_kernel_3_and_3_conv2d.evaluate(x_train,y_train)
```
